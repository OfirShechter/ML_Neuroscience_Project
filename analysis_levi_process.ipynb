```python
old_val = pc_pymat[10000]
curr_val = 0
process = [old_val]
for i in range(0,72000000,50):
    curr_val = pc_pymat[i]
    diff = curr_val - old_val
    process.append(diff)
    old_val = curr_val
```

```python
plt.figure(figsize=(15,5))
plt.plot(process[600000:700000])
plt.plot(process)
```
